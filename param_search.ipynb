## CEM

### Experiment fusion

Join first all the experiments in a single dataframe to evaluate them

```python
l = []
nn_l = []
svm_l = []
d = {}
algo_chosen = 'cem'
for file_name in algo_files[algo_chosen]:
    print(file_name)
    db, bb, algo = parse_file_name(file_name)
    df = pd.read_csv(join(mypath, file_name), names=HEADERS[algo])
    max_cf_ratio = df.cf_ratio.max()
    df['algo'] = algo
    df['bb'] = bb
    df['db'] = db
    l.append(df)
    if bb == 'NN':
        nn_l.append(df)
    else:
        svm_l.append(df)
    d[(algo, bb, db)] = df

df = pd.concat(l, axis=0, ignore_index=True)
```

### Scoring

We fix a resolution to get all the executions within a range. Then we get the average of the ratio of found counterfactuals and the number of datasets that have at least one element within range

```python
def get_best_cem_conf(l):
    resolution = {
        'kappa': 0.05,
        'beta': 0.05,
        'c_init': 1,
        'c_steps': 1
    }

    all_confs = []

    for df in l:
        for conf in df.iterrows():
            # print('New conf')
            score = 1
            hits = 0
            for ndf in l:
                nndf = ndf.loc[
                    (ndf.kappa >= conf[1]['kappa'] - resolution['kappa']) & 
                    (ndf.kappa <= conf[1]['kappa'] + resolution['kappa']) &
                    (ndf.beta >= conf[1]['beta'] - resolution['beta']) & 
                    (ndf.beta <= conf[1]['beta'] + resolution['beta']) & 
                    (ndf.c_init >= conf[1]['c_init'] - resolution['c_init']) & 
                    (ndf.c_init <= conf[1]['c_init'] + resolution['c_init']) & 
                    (ndf.c_steps >= conf[1]['c_steps'] - resolution['c_steps']) & 
                    (ndf.c_steps <= conf[1]['c_steps'] + resolution['c_steps']) 

                ]
                score += nndf.cf_ratio.mean()
                hits += int(nndf.cf_ratio.mean() > 0)
            all_confs += [((conf[1]['kappa'], conf[1]['beta'], conf[1]['c_init'], conf[1]['c_steps']), score, hits)]
            # print(score, hits)
            # print('---------------------')
    best_cem_conf = sorted(all_confs, key=lambda a: (a[2], a[1]), reverse=True)[0][0]
    return best_cem_conf
```

Finally, we sort by score and hits, and we get the best one to get our configuration

```python
best_cem_conf = get_best_cem_conf(l)
print(f'The best general CEM configuration is {best_cem_conf}')
```

```python
best_cem_conf = get_best_cem_conf(nn_l)
print(f'The best CEM configuration for NN is {best_cem_conf}')
```

```python
best_cem_conf = get_best_cem_conf(svm_l)
print(f'The best CEM configuration for SVM is {best_cem_conf}')
```
