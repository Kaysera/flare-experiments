```python
import numpy as np
import pandas as pd
import random

import matplotlib.pyplot as plt
```

```python
seed = 42

random.seed(seed)
np.random.seed(seed)
```

```python
# Get a numpy array with 10000 random numbers from a normal distribution centered in 30 and standard deviation 5

data_age = np.random.normal(30, 5, 10000)

# Plot a histogram of the data
# with title: 'age' and the standard deviation as the x-axis label
# and y-axis label: 'frequency'
# with font size 20

plt.figure(figsize=(10, 7))

plt.rcParams.update({'font.size': 25})

plt.ylabel('Frequency')

plt.title('Age')
plt.xlabel(f'std = 5')

plt.hist(data_age, 50)
```

```python
# Get a numpy array with 10000 random numbers from a normal distribution centered in 80 and standard deviation 10

data_weight = np.random.normal(80, 10, 10000)
data_weight_cut = data_weight[data_weight>65]
data_weight_cut = data_weight_cut[data_weight_cut<100]

# Plot a histogram of the data

plt.figure(figsize=(10, 7))

plt.rcParams.update({'font.size': 25})

plt.ylabel('Frequency')

plt.title('Weight')

plt.xlabel(f'std = 10')

plt.hist(data_weight_cut, 50)
```
